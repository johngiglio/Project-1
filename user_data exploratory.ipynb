## Identify Highly Active Users

Create total columns for Votes Sent and Compliments Received. Eliminate users who have no votes, compliments, reviews, fans.

```python
elite_users = user_data[user_data['elite'] != 'None' ]
print(f"{'{:.2f}'.format(100*len(elite_users)/len(user_data))}% of the users in this dataset are Elite.")
```

```python
print(f"Elite users have written {'{:.2f}'.format(100*sum(elite_users['review_count'])/sum(user_data['review_count']))}% of the total number of reviews written by this group of users.")
```

The elite column contains the years that each user was Elite in string format. It needs to be split into a list to be counted.

```python
elite_users['# elite years'] = elite_users['elite'].apply(lambda x: len(x.split(sep = ', ')))
elite_users.head()
```

Min review_count for elite users is 1, find a minimum review count threshold to reduce number of elite users.

```python
elite_users[['review_count','# elite years']].describe()
```

Eliminate users that have written less than 50 reviews from the elite group.

```python
len(elite_users[elite_users['review_count'] < 50])
```

```python
elite_users = elite_users[elite_users['review_count'] >= 50]
```

```python
print(f"Elite users now make up {'{:.2f}'.format(elite_users['review_count'].count()/user_data['review_count'].count()*100)}% of the population and wrote {'{:.2f}'.format(elite_users['review_count'].sum()/user_data['review_count'].sum()*100)}% of the reviews")
```

```python
plt.figure(figsize=(20,10))
plt.xlim(0,14)
plt.ylim(0,4000)
plt.scatter(elite_users['# elite years'],elite_users['review_count'], alpha = 0.15, s =elite_users['fans'], c = elite_users['# elite years'] )

plt.show()
```
