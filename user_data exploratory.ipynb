## Identify Highly Active Users

Create total columns for Votes Sent and Compliments Received. Eliminate users who have no votes, compliments, reviews, fans. t-tests for each city/region 

Yelp has already identified highly active/effective users in the data set in the elite column. Split the data set into elite and row subsets.

```python
set(user_data['elite'])
```

```python
elite_users = user_data[user_data['elite'] != 'None' ]
len(elite_users)/len(user_data)
```

The elite column contains the years that each user was Elite in string format. It needs to be split into a list to be counted.

```python
elite_users['# elite years'] = elite_users['elite'].apply(lambda x: len(x.split(sep = ', ')))
```

```python
elite_users.head()
```

```python
elite_users[['review_count','# elite years']].describe()
```

```python
len(elite_users[elite_users['review_count'] < 10])
```

```python
elite_users = elite_users[elite_users['review_count'] >= 50]
```

```python
elite_users = elite_users[elite_users['# elite years'] >= 2]
```

```python
print(f"Elite users make up {'{:.2f}'.format(elite_users['review_count'].count()/user_data['review_count'].count()*100)}% of the population and wrote {'{:.2f}'.format(elite_users['review_count'].sum()/user_data['review_count'].sum()*100)}% of the reviews")
```

```python
plt.figure(figsize=(20,10))
plt.xlim(0,14)
plt.ylim(0,4000)
plt.scatter(elite_users['# elite years'],elite_users['review_count'], alpha = 0.15, s =elite_users['fans'], c = elite_users['# elite years'] )

plt.show()
```
